```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import pickle



columns=['actor_1_name','actor_2_name','actor_3_name','director_name','genres','movie_title','combine']


data[columns].isnull().values.any()#no null values

def get_important_features(data):
    important_features=[]
    for i in range (0,data.shape[0]):
        important_features.append(data['combine'][i])
    return important_features


#creating a column to hold the combined strings
data['important_features']=get_important_features(data)
data
```

```python
data['important_features']
```

```python
imp = pd.DataFrame(data['important_features'])
imp
```

```python
tfidf = TfidfVectorizer(stop_words='english')
#data['Description'] = data['Description'].fillna('')
tfidf_matrix = tfidf.fit_transform(data['important_features'])
tfidf_matrix.shape
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(data.index, index=data['movie_title']).drop_duplicates()
indices
```

```python
def get_recommendations(movie_title, cosine_sim=cosine_sim):
    idx = indices[movie_title]
    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))
    # Sort the movies based on the similarity score
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    movie_indices = [i[0] for i in sim_scores]
    # Return the top 5 most similar movies
    movies=data['movie_title'].iloc[movie_indices]
    id=data['movie_id'].iloc[movie_indices]
    dict={"Movies":movies,"id":id}
    final_df=pd.DataFrame(dict)
    final_df.reset_index(drop=True,inplace=True)
    return final_df


get_recommendations('Avatar')
get_recommendations('Spider-man: far from home')
```
